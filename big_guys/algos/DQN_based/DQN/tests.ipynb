```python
from RL.fa_PG.algos.DQN_based.DQN.DQN import DQN
import matplotlib.pyplot as plt
```

```python
%%time
stats, agent = DQN("Berzerk-v0", epochs=5, 
                   eps_sched={"start": 0.9, "end": 0.05, "decay": 200})

plt.plot(stats.EpRet)
plt.show()
plt.plot(stats.EpLen)
plt.show()
plt.plot(agent.loss)
plt.show()
plt.plot(agent.eps)
plt.show()
```

```python
%%time
stats, agent = DQN("Berzerk-v0", epochs=5, 
                   eps_sched={"start": 0.9, "end": 0.05, "decay": 200},
                   double=True)

plt.plot(stats.EpRet)
plt.show()
plt.plot(stats.EpLen)
plt.show()
plt.plot(agent.loss)
plt.show()
plt.plot(agent.eps)
plt.show()
```

```python
%%time
stats, agent = DQN("Berzerk-v0", epochs=5, 
                   eps_sched={"start": 0.9, "end": 0.05, "decay": 200},
                   dueling=True, start_steps=100)

plt.plot(stats.EpRet)
plt.show()
plt.plot(stats.EpLen)
plt.show()
plt.plot(agent.loss)
plt.show()
plt.plot(agent.eps)
plt.show()
```
